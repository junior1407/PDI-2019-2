### Adding noise to an image

```python
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np

imfile = '../db/lena.png'
```

```python
img = cv2.imread(imfile, cv2.IMREAD_COLOR)
noise = np.zeros(img.shape, img.dtype)
noise2 = np.random.random(img.shape)
cv2.randn(noise, 0, 150)
img2 = img + noise

plt.figure(figsize= (16,6))
plt.subplot(131), plt.title("Img1"), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(132), plt.title("Noise"), plt.imshow(noise)
plt.subplot(133), plt.title("Img2"), plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()

```

## Salt and pepper

```python
img = cv2.imread(imfile, cv2.IMREAD_COLOR)
noise = np.zeros((img.shape[0], img.shape[1]), img.dtype)
cv2.randu(noise, 0, 255)

density = 20
salt = noise > 255 - density
pepper = noise < 5

img2 = img.copy()
img2[salt == True] = 255
img2[pepper == True] = 0

plt.figure(figsize= (16,7))
plt.subplot(121), plt.title("Img1"), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(122), plt.title("Img2"), plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

## median filter

```python
imgFiltered = cv2.medianBlur(img2, 5)

plt.imshow(cv2.cvtColor(imgFiltered , cv2.COLOR_BGR2RGB))
plt.show()
```

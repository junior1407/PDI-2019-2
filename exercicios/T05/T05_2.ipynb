# Salt and Pepper

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def addSaltnPepper(data, saltiness, spiceness):
    height,width = data.shape[:2]
    noise = np.zeros((height,width), data.dtype)
    cv2.randu(noise,0,255)        
    salt = noise > saltiness
    pepper = noise < spiceness
    
    img2 = data.copy()
    img2[salt] = 255
    img2[pepper] = 0
    return img2
```

```python
imgfile = '../../db/lena.png'
img= cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)
height,width = img.shape[:2]
```

```python
#Adds salt'n pepper noise
noisy = addSaltnPepper(img, 250,8)
```

```python
plt.figure(figsize=(6,3))
plt.subplot(121), plt.title('Original'), plt.imshow(img, cmap='gray')
plt.subplot(122), plt.title('With noise'), plt.imshow(noisy, cmap='gray')
plt.show()
```

## Limiarização (Faixa)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imgfile = '../../db/cameraman_original.png'  #Image
img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)

#v1 < p < v2 
v1 = 200
v2 = 240

mask = np.multiply(img > v1, img < v2)
img2 = np.multiply(mask,img)

plt.figure(figsize = (6,6))
plt.subplot(121), plt.title("Original"),plt.imshow(img, cmap='gray')
plt.subplot(122), plt.title("Depois do Limiar"), plt.imshow(img2, cmap='gray')
plt.show()
```

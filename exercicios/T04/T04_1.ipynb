## Limiarização (quase Binarização)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imgfile = '../../db/cameraman_original.png'  #Image
imgOriginal = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)

limit= 120
img = imgOriginal > 120

plt.figure(figsize = (6,6))
plt.subplot(121), plt.title("Original"),plt.imshow(imgOriginal, cmap='gray')
plt.subplot(122), plt.title("Depois do Limiar"), plt.imshow(img, cmap='gray')
plt.show()
```

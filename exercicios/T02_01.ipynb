```python
#Imports
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Image Config
height = 100
width = 100
n = 100
x = np.linspace(-height//2, height//2, n)
y = np.linspace(-width//2, width//2,n)
X, Y = np.meshgrid(x,y)
```

```python
a = 1
b = 2
c = 3
img = ((X/a)**2 + (Y/b)**2)*c
plt.imshow(img, cmap = 'gray')
plt.show()
```

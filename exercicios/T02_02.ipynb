# Geração do seno com Inclinação
### Create the image of a rotated sin using rotation of coordinates.

```python
#Imports
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Image Config
height = 100
width = 100
n = 100
x = np.linspace(-height//2, height//2, n)
y = np.linspace(-width//2, width//2,n)
X, Y = np.meshgrid(x,y)
```

```python
rotation_degrees= 30; #Rotate 30 degrees clockwise
rotation_radian = (np.pi/180)*rotation_degrees

Xr = np.cos(rotation_radian)*X + np.sin(rotation_radian)*Y
Yr = -np.sin(rotation_radian)*X + np.cos(rotation_radian)*Y

x_y = np.sin(Xr)
plt.imshow(x_y, cmap = 'gray')
plt.show()
```

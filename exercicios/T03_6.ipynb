### Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def applyMask(img, mask):
    height,width = img.shape[:2]
    blank_image = np.zeros((height,width,3), np.uint8)
    for i in range(height):
        for j in range(width):
            if (mask[i][j]):
                blank_image[i][j] = img[i][j]
    return blank_image

def resize(img, width, height):
    return cv2.resize(img,(width, height))
```

```python
imgfile = '../db/faces.jpg'
maskfile = '../db/rect.png'
img = cv2.imread(imgfile)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)     
height,width = img.shape[:2]

mask = cv2.imread(maskfile, cv2.IMREAD_GRAYSCALE)
mask_resized=cv2.resize(mask,(width, height))
masked = applyMask(img, mask_resized)
```

```python
plt.figure(figsize=(20,60))
plt.subplot(131)
plt.imshow(img)

plt.subplot(132)
plt.imshow(mask_resized, cmap="gray")

plt.subplot(133)
plt.imshow(masked)
```

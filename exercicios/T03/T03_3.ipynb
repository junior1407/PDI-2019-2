# Rotação
### Implement image rotation using Numpy and OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotate(img, degrees):
    height,width = img.shape[:2]
    T=cv2.getRotationMatrix2D((width/2, height/2),degrees,1)
    return cv2.warpAffine(img, T, (width, height))
```

```python
imfile = '../db/lena.png'
img = cv2.imread(imfile)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)       
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)

plt.subplot(122)
degrees= 20
plt.title("Rotated " + str(degrees) + " degrees")

img_rotated = rotate(img,degrees)
plt.imshow(img_rotated)
```

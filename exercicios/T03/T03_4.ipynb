# Rescalonamento
### Implement image resizing using OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def resizeByScale(img, scale):
    return cv2.resize(img, None, fx = scale, fy = scale)

def resize(img, width, height):
    return cv2.resize(img,(width, height))
```

```python
imfile = '../db/lena.png'
img = cv2.imread(imfile)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)       
plt.figure(figsize=(8,8))
plt.subplot(221)
plt.title("Original")
plt.imshow(img)

plt.subplot(222)
plt.title("Scaled by 2")
img2 = resizeByScale(img,2)
plt.imshow(img2)

plt.subplot(223)
plt.title("Original")
plt.imshow(img)


plt.subplot(224)
plt.title("Scaled down to w= 250,h = 150")
img3 = resize(img, 250,150)
plt.imshow(img3)
```

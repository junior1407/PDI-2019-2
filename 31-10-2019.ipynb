```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
imfile = '../db/lena.png'
img = cv2.imread(imfile)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

height, width = img.shape[:2]
print("Height = " + str(height))
print("Width  = " + str(width ))
img2 = img+70
plt.figure(1, figsize = (16,6))
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(img2)
```

```python
imfile1= '../db/lena.png'
imfile2 = '../db/astronaut.png'
img1 = cv2.imread(imfile1).astype(np.float32)/255
img2 = cv2.imread(imfile2).astype(np.float32)/255
alpha = 0.7
img3 = alpha*img1 + (1-alpha) * img2
img4 = cv2.max(img1,img2  #absdiff
plt.figure(3, figsize = (16,6))
plt.subplot(1,4,1); plt.title("Img1"); plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.subplot(1,4,2); plt.title("Img2"); plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.subplot(1,4,3); plt.title("Img3"); plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.subplot(1,4,4); plt.title("Img4"); plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
plt.show()
```
